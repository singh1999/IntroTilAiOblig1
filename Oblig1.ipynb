```python
dataframe["Dato"] = panda.to_datetime(dataframe.Dato, format="%d/%m/%Y")
dataframe["Måned"] = dataframe["Dato"].dt.month
dataframe["Dag"] = dataframe["Dato"].dt.day_name()
```

```python
dataframe.head()
```

```python
dataframe.dropna(subset = ["Dag"], inplace=True)
```

```python
plt.hist(dataframe.Dag, bins=7, rwidth=0.98)
plt.xlabel('Dag')
plt.ylabel('Count')
plt.show()
```

```python
encoded_columns = panda.get_dummies(dataframe['Fylke'])
dataframe = dataframe.join(encoded_columns).drop('Fylke', axis=1)
```

```python
dataframe.sample(10)
```

```python
datatime = dataframe.Tidspunkt_Faktisk_Avgang_Holdeplass_Fra.str.split(":").map(lambda x: x[0])
dataframe["Avgang_Time"] = datatime
```
